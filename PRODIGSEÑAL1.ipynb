```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
frecuencia_señal = 2  # Frecuencia de la señal (Hz)
duración = 1          # Duración de la señal (s)
amplitud = 1          # Amplitud de la señal

# Sample rates
fs1 = 10  # Frecuencia de muestreo 1
fs2 = 100 # Frecuencia de muestreo 2

# Tiempo para cada frecuencia de muestreo
t1 = np.linspace(0, duración, int(fs1 * duración), endpoint=False)
t2 = np.linspace(0, duración, int(fs2 * duración), endpoint=False)

# Generación de señales senoidales
señal1 = amplitud * np.sin(2 * np.pi * frecuencia_señal * t1)
señal2 = amplitud * np.sin(2 * np.pi * frecuencia_señal * t2)

# Transformada de Fourier de ambas señales
fft1 = np.fft.fft(señal1)
fft2 = np.fft.fft(señal2)

# Cálculo de las frecuencias asociadas
frecuencias1 = np.fft.fftfreq(len(fft1), d=1/fs1)
frecuencias2 = np.fft.fftfreq(len(fft2), d=1/fs2)

# Espectro de potencia
potencia1 = np.abs(fft1)**2
potencia2 = np.abs(fft2)**2

# Graficar señales en el tiempo
plt.figure(figsize=(14, 10))

# Señales en el dominio del tiempo
plt.subplot(2, 2, 1)
plt.stem(t1, señal1, linefmt='b-', markerfmt='bo', basefmt=" ", label=f'Sample rate: {fs1} Hz')
plt.title('Señal (Sample rate: 10 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.stem(t2, señal2, linefmt='r-', markerfmt='rx', basefmt=" ", label=f'Sample rate: {fs2} Hz')
plt.title('Señal (Sample rate: 100 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()

# Espectros de potencia
plt.subplot(2, 1, 2)
plt.stem(frecuencias1, potencia1, linefmt='b-', markerfmt='bo', basefmt=" ", label=f'Espectro (Sample rate: {fs1} Hz)')
plt.stem(frecuencias2, potencia2, linefmt='r-', markerfmt='rx', basefmt=" ", label=f'Espectro (Sample rate: {fs2} Hz)')
plt.title('Espectro de Potencia de ambas señales')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Potencia')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
